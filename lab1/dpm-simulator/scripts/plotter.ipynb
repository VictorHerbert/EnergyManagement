```python
import sys
import pandas as pd
from matplotlib import pyplot as plt
```

```python
IN_FILE = '../build/w1_timeout.log'
OUT_FILE = '../build/w1_timeout.png'
```

```python
df = pd.read_csv(IN_FILE)
df['energy_save_percent'] = (df['energy_without_dpm'] - df['energy_with_dpm'])/df['energy_without_dpm']
df.head()
```

```python
plt.grid()
plt.xlabel('Timeout (ms)')
plt.ylabel('Energy Save (%)')

plt.fill_between([0,4], [df['energy_save_percent'].max(), df['energy_save_percent'].max()], df['energy_save_percent'].min(), label='Region 1', color='lightblue')
plt.fill_between([4,20], [df['energy_save_percent'].max(), df['energy_save_percent'].max()], df['energy_save_percent'].min(), label='Region 2', color='green', alpha=.5)

plt.legend()
plt.plot(df['timeout'], df['energy_save_percent'], color='black')

plt.savefig(OUT_FILE)
```
